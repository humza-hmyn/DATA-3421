```python
#	8. Image Rotation:
#●	Rotate the 'Pandas.png' image by 45 degrees without cropping the corners and display the result.
#●	Discuss the challenges and solutions in rotating images like 'Pandas.png'

height, width = img.shape[:2]
rotation_matrix = cv2.getRotationMatrix2D((width / 2, height / 2), 45, 1)
rotated_img = cv2.warpAffine(img, rotation_matrix, (width, height))
plt.imshow(rotated_img)
#uh oh the corners are cropped0
```

```python
rot_image = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
plt.imshow(rot_image)
```

```python
#	9. Image Flipping:
#●	Flip the 'Pandas.png' image horizontally and vertically, displaying both results.
#●	Explain when flipping an image might be useful in preprocessing steps.

hori_flip= cv2.flip(img, 1)
vert_flip= cv2.flip(img, 0)

plt.subplot(1, 2, 1)
plt.imshow(hori_flip)

plt.subplot(1, 2, 2)
plt.imshow(vert_flip)
```

Flipping images help train your model to work with flipped images - essentially by outlining and feeding in detailed flipped images to an algorithim, you're essentially teaching it how to analyze data that may be simple to comprehend for us humans. Lines of code cannot tell if a picture it just analyzed is the same thing just upside down like we can.

```python
#10. Edge Detection:
#●	Apply the Sobel filter to find edges in the 'Pandas.png' image and display the result. 
#Explain how it computes the gradients.
img_gs = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

sobelx = cv2.Sobel(img_gs, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(img_gs, cv2.CV_64F, 0, 1, ksize=3)
sobel_alpha = cv2.magnitude(sobelx, sobely)

plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.imshow(sobelx, cmap='gray')

plt.subplot(1, 3, 2)
plt.imshow(sobely, cmap='gray')

plt.subplot(1, 3, 3)
plt.imshow(sobel_alpha, cmap='gray')
```
